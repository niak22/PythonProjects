```python
import matplotlib.pyplot as plt

# Define the data
stages = ["Problem Identification", "Problem Definition", "Solution Generation", "Solution Implementation", "Solution Evaluation"]
engagement_levels = [("Students", "Medium"), ("Educators", "High"), ("Mentors", "Medium"), ("Employers", "Low"), ("Investors", "High")]

# Create a dictionary of stakeholder colors
colors = {"Students": "blue", "Educators": "green", "Mentors": "yellow", "Employers": "orange", "Investors": "red"}

# Create the graph
fig, ax = plt.subplots()

# Plot dummy lines for each stakeholder to create entries in the legend
for stakeholder, level in engagement_levels:
    plt.plot([], [], color=colors[stakeholder], label=stakeholder)

# Plot the lines for each stakeholder with numerical y-values and a small offset
offset = 0.2  # Adjust the offset as needed
for i, (stakeholder, level) in enumerate(engagement_levels):
    color = colors[stakeholder]
    # Assign numerical values to engagement levels for plotting
    y_value = {"Low": 1, "Medium": 2, "High": 3}[level]
    plt.plot(stages, [y_value + i * offset] * len(stages), color=color)

# Add labels and title
plt.xlabel("Stages of the Problem-Solving Process")
plt.ylabel("Engagement Level")
plt.title("Engagement of Stakeholders in the Problem-Solving Process")


# Rotate x axis labels
plt.xticks(rotation=45)

# Add grid lines
plt.grid(True)

# Adjust y axis limits
plt.ylim(-0.5, 3.5)

# Add legend
plt.legend()

# Show the graph
plt.show()

```
